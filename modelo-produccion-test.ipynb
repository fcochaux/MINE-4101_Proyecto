### mover a tabla de samuel

```python
data_df = spark.read \
    .format("jdbc") \
    .option("url", url) \
    .option("query", query3) \
    .option("user", username) \
    .option("password", password) \
    .option("TrustServerCertificate",True) \
    .option("driver", driver_mssql) \
    .load().dropna().withColumnRenamed('llavePuntoConsumo','lpc')

data_df.show()
```

```python
predictions_df = spark.read \
    .format("jdbc") \
    .option("url", url) \
    .option("dbtable", '[DWSRBI_KRONOS].[prueba].[predicciones]') \
    .option("user", username) \
    .option("password", password) \
    .option("TrustServerCertificate",True) \
    .option("driver", driver_mssql) \
    .load().select(col('LlavePuntoConsumo'),col('prediction'))
```

```python
result_join = predictions_df\
.join(data_df, predictions_df.LlavePuntoConsumo == data_df.lpc, 'inner')\
.select(col('LlavePuntoConsumo').alias('llavePuntoConsumo'),col('prediction').alias('porcentajePosibleFraude')\
        ,col('CodigoCliente').alias('nis_rad'))

result_join = result_join.withColumn('identificadorProceso',concat(lit('15-SEP-2022')))\
.withColumn('esPosibleFraude',F.when((F.col("porcentajePosibleFraude") > 0.99) , 1).otherwise(0) )
#df.show()


result_join.toPandas().head(100)
```

```python
result_join.filter(col('esPosibleFraude')==1).show()
```

```python
result_join.write.format("jdbc") \
    .mode('append') \
    .option("url", url) \
    .option("dbtable", '[DWSRBI_KRONOS].[Hecho].[ProbabilidadFraude]') \
    .option("user", username) \
    .option("password", password) \
    .option("TrustServerCertificate",True) \
    .option("driver", driver_mssql) \
    .save()
```
